```python
meta = pd.DataFrame()
meta['nullity'] = df.isna().sum()
meta['pct_null'] = meta['nullity']*100/df.shape[0]
meta['dtype'] = df.dtypes
```

```python
meta
```

```python
dict_col = {'columns':meta.index.tolist(), 'expect_datatype':[], 'related_to':[], 'relationship_strength':[]}
for i in meta.index:
    print(f"{i} should have")
    print("Expected data type:   int  ordinal  nominal   temporal    none")
    expected_dtype = input("expected data type")
    dict_col['expect_datatype'].append(expected_dtype)
    print("is it related to - house    peripheral   location")
    relation = input("is it related to house, perpheral or location")
    dict_col['related_to'].append(relation)
    print("strength     strong  weak    medium")
    strength = input('what to do you think about the realtionship with target?')
    dict_col['relationship_strength'].append(strength)
    print("----------------------------------------")
    print("\n\n")
```

```python
col_df = pd.DataFrame(dict_col)
col_df
```

```python
col_df.index = col_df['columns']
meta_df = pd.concat([meta, col_df], axis=1)
meta.shape, col_df.shape, meta_df.shape
```

```python
meta_df.to_csv('../data/meta_data.csv', index=False)
```
